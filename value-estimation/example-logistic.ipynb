```python
from __future__ import print_function
from vowpalwabbit import pyvw
```

```python
import numpy as np
import numpy.random as rnd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Generate the model
N = 1000
F = 10
W = rnd.randn(F, 1)
b = -3 + rnd.randn()
```

```python
# Generate sample input
def sigmoid(x):
    return 1/(1  + np.exp(-x))

def log_loss(y, p):
    return -(y * np.log(p) + (1-y) * np.log(1-p))

X = rnd.randn(N, F)
z = np.dot(X, W) + b
y = rnd.binomial(1, sigmoid(z)).reshape(N, 1)

print("mean value: {}".format(np.mean(y>0)))
```

```python
plt.hist(y)
plt.title('histogram of responses')
```

```python
# Convert sample into VW input
def generate_vw(X, y):
    y[y[:,0] <= 0, 0] = -1
    samples = []
    for i in xrange(X.shape[0]):
        example = "{} |".format(y[i, 0])
        for f in xrange(X.shape[1]):
            example += " {}:{}".format(f, X[i,f])
        samples.append(example)
    return samples
generate_vw(X[0:2,:],y[0:2])
```

```python
# Run stochastic gradient descent
ITER = 100
vw = pyvw.vw('--loss_function logistic --link logistic -l 0.1 -b 20') 
samples = generate_vw(X,y)
iter_loss = []
predictions = []
for iteration in xrange(ITER):
    predictions = []
    loss  = 0
    for i in xrange(len(samples)):
        ex = vw.example(samples[i])
        ex.learn()
        predictions.append((vw.predict(ex)))
        loss += log_loss(y[i, 0], (vw.predict(ex)))
    iter_loss.append(loss/N)
```

```python
plt.plot(iter_loss)
```

```python
plt.scatter(y[:,0], predictions)
plt.title('Predictions vs. Actual')
```

```python
What = [vw.get_weight(i) for i in xrange(F)]
plt.scatter(W[:, 0], What)
plt.title('Predicted vs. Actual weight')
```
